# Prompt : templating et chaining

## Prompt Template

```python
from langchain import PromptTemplate

template="Tu es le consultant d'une nouvelle compagnie. Quels sont les noms qu'on peut donner à une compagnie qui vend des {produit}? "

prompt=PromptTemplate.from_template(template)

print(prompt)
```

```python
print(prompt.format(produit="café"))
```

## LLM Chain

```python
from langchain.chains import LLMChain
```

```python
chain=LLMChain(llm=llm, prompt=prompt)
```

```python
print(chain.run("café"))
```

On peut avoir plusieurs variables d'entrées. Dans ce cas, il faut utliser un dictionnaire lors de chain.run()

# Simple Sequential chain

```python
from langchain.chains import SimpleSequentialChain
```

```python
template2="Quel est le meilleur titre accrocheur pour un livre qui parle de {thème}"

prompt2=PromptTemplate.from_template(template2)
chain2=LLMChain(llm=llm, prompt=prompt2)
chain2.run("sentiment")
```

```python
template3="génère une idée d'illustration pour un livre qui porte le titre {titre} "

prompt3=PromptTemplate.from_template(template3)
chain3=LLMChain(llm=llm, prompt=prompt3)
```

```python
global_chain=SimpleSequentialChain(chains=[chain2,chain3], verbose=True)

print(global_chain.run("dieu"))
```

```python
import speech_recognition as sr


def transcrire_audio_en_yoruba(chemin_audio):
    # Créer un objet Recognizer
    recognizer = sr.Recognizer()


    # Charger le fichier audio
    with sr.AudioFile(chemin_audio) as source:
        audio = recognizer.record(source)  # Enregistrer l'audio



    # Utiliser Google Web Speech API pour la transcription en yoruba
    try:
        texte_transcrit = recognizer.recognize_google(audio, language='yo-NG')  # Transcription en yoruba
        return texte_transcrit
    except sr.UnknownValueError:
        return "Impossible de transcrire l'audio : le langage n'a pas pu être identifié"
    except sr.RequestError as e:
        return f"Erreur lors de la requête vers l'API Google : {e}"

# Exemple d'utilisation de la fonction
chemin_audio = 'WhatsApp Audio 2024-05-14 à 10.39.58_a42bb18c.opus'

resultat_transcription_yoruba = transcrire_audio_en_yoruba(audio_file_path)
print("Résultat de la transcription en yoruba :")
print(resultat_transcription_yoruba)
```
